```python
# 3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution. 

def bernoulli_pmf(x, p):
    """
    Computes the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The value of the random variable (0 or 1).
    p (float): The probability of success (0 ‚â§ p ‚â§ 1).

    Returns:
    float: Probability P(X = x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.6  # Probability of success
print("P(X=0):", bernoulli_pmf(0, p))
print("P(X=1):", bernoulli_pmf(1, p))
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a non-normal (exponential) distribution
np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)  # Exponential distribution

# Step 2: Function to perform CLT simulation
def central_limit_theorem(population, sample_size, num_samples=1000):
    sample_means = []  # Store sample means

    # Step 3: Draw samples and compute means
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    # Step 4: Plot histogram of sample means
    plt.figure(figsize=(8, 5))
    sns.histplot(sample_means, bins=30, kde=True, color='blue', edgecolor='black')

    # Add mean and normal curve
    plt.axvline(np.mean(sample_means), color='red', linestyle='--', label=f'Mean={np.mean(sample_means):.2f}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title(f'CLT Demonstration (Sample Size = {sample_size})')
    plt.legend()
    plt.grid(alpha=0.5)
    
    # Show plot
    plt.show()

# Step 5: Test CLT with different sample sizes
for size in [5, 30, 100]:  # Small, medium, large sample sizes
    central_limit_theorem(population, sample_size=size)
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Performs a one-sample Z-test for a given sample against a population mean.
    
    Parameters:
    sample (list or np.array): The sample dataset.
    population_mean (float): The known population mean.
    population_std (float): The known population standard deviation.
    alpha (float): The significance level (default: 0.05).

    Returns:
    None (prints hypothesis test results).
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    
    # Compute Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Compute P-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.5f}")

    # Decision rule
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (H0) - Significant difference found.")
    else:
        print("Conclusion: Fail to reject the null hypothesis (H0) - No significant difference.")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Simulated sample
z_test(sample_data, population_mean=50, population_std=10)
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a given sample dataset.

    Parameters:
    sample (list or np.array): The sample dataset.
    confidence (float): The confidence level (default: 0.95 for 95% CI).

    Returns:
    tuple: (lower_bound, upper_bound) of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)
    n = len(sample)

    # Critical Z-value (for large samples) or t-value (if sample size is small)
    z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)  # For normal distribution

    # Compute margin of error
    margin_of_error = z_critical * (sample_std / np.sqrt(n))

    # Confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"{confidence * 100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Simulated sample dataset
confidence_interval(sample_data)
```

```python
# 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for a given sample dataset using the t-distribution.
    
    Parameters:
    sample (list or np.array): The sample dataset.
    confidence (float): The confidence level (default: 0.95 for 95% CI).
    
    Returns:
    tuple: (lower_bound, upper_bound) of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)
    n = len(sample)

    # Critical t-value (used for small samples or unknown population std)
    t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df=n - 1)  # Two-tailed test

    # Compute margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"{confidence * 100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return lower_bound, upper_bound

# Step 1: Generate Data from a Normal Distribution
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Normal(mean=50, std=10, size=30)

# Step 2: Compute and Interpret Confidence Interval
calculate_confidence_interval(sample_data)
```

```python
# 24. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def simulate_uniform_distribution(a=0, b=10, num_samples=1000):
    """
    Simulates a continuous uniform distribution, plots the histogram, and calculates the expected value.

    Parameters:
    a (float): Lower bound of the uniform distribution.
    b (float): Upper bound of the uniform distribution.
    num_samples (int): Number of random samples to generate.

    Returns:
    None (Displays the plot and prints the expected value).
    """
    # Generate random samples
    samples = np.random.uniform(a, b, num_samples)

    # Compute the expected value (theoretical and empirical)
    expected_value_theoretical = (a + b) / 2
    expected_value_empirical = np.mean(samples)

    # Plot histogram
    plt.figure(figsize=(8, 5))
    plt.hist(samples, bins=30, density=True, alpha=0.6, color='blue', label='Sampled Data')

    # Plot the uniform PDF for reference
    x = np.linspace(a, b, 100)
    pdf_values = uniform.pdf(x, loc=a, scale=(b-a))
    plt.plot(x, pdf_values, color='red', label='Uniform PDF')

    # Labels and title
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'Continuous Uniform Distribution U({a}, {b})')
    plt.legend()
    plt.grid(alpha=0.5)

    # Show the plot
    plt.show()

    # Print expected values
    print(f"Theoretical Expected Value: {expected_value_theoretical:.2f}")
    print(f"Empirical Expected Value (from samples): {expected_value_empirical:.2f}")

# Example usage
simulate_uniform_distribution(a=2, b=12, num_samples=1000)
```

```python
# 25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(dataset1, dataset2, label1="Dataset 1", label2="Dataset 2"):
    """
    Compares the standard deviations of two datasets and visualizes the difference.

    Parameters:
    dataset1 (list or np.array): First dataset.
    dataset2 (list or np.array): Second dataset.
    label1 (str): Label for first dataset.
    label2 (str): Label for second dataset.

    Returns:
    None (Displays a histogram and prints standard deviations).
    """
    # Compute standard deviations
    std_dev1 = np.std(dataset1, ddof=1)  # Sample standard deviation
    std_dev2 = np.std(dataset2, ddof=1)

    # Print standard deviations
    print(f"{label1} - Standard Deviation: {std_dev1:.2f}")
    print(f"{label2} - Standard Deviation: {std_dev2:.2f}")

    # Plot histograms
    plt.figure(figsize=(8, 5))
    plt.hist(dataset1, bins=20, alpha=0.6, color='blue', label=f"{label1} (œÉ={std_dev1:.2f})", edgecolor='black')
    plt.hist(dataset2, bins=20, alpha=0.6, color='red', label=f"{label2} (œÉ={std_dev2:.2f})", edgecolor='black')

    # Labels and title
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Comparison of Standard Deviations')
    plt.legend()
    plt.grid(alpha=0.5)

    # Show the plot
    plt.show()

# Example usage
np.random.seed(42)
dataset1 = np.random.normal(loc=50, scale=10, size=1000)  # Normal(mean=50, std=10)
dataset2 = np.random.normal(loc=50, scale=20, size=1000)  # Normal(mean=50, std=20)

compare_standard_deviations(dataset1, dataset2, label1="Dataset 1 (œÉ=10)", label2="Dataset 2 (œÉ=20)")
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_range_and_iqr(data):
    """
    Calculates and prints the range and interquartile range (IQR) of a dataset.
    
    Parameters:
    data (list or np.array): The dataset.

    Returns:
    None (Prints the range and IQR, and displays a boxplot).
    """
    # Calculate range
    data_range = np.max(data) - np.min(data)

    # Calculate quartiles
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    
    # Compute IQR
    IQR = Q3 - Q1

    # Print results
    print(f"Range: {data_range:.2f}")
    print(f"Interquartile Range (IQR): {IQR:.2f}")

    # Plot boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(data, color='lightblue')
    plt.title("Boxplot of Dataset (Showing IQR)")
    plt.xlabel("Values")
    plt.grid(alpha=0.5)
    plt.show()

# Generate a normal distribution dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=500)  # Normal(mean=50, std=15, size=500)

# Compute and display range and IQR
calculate_range_and_iqr(data)
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_skewness_kurtosis(data):
    """
    Calculates and prints the skewness and kurtosis of a dataset.
    Also visualizes the dataset using a histogram.

    Parameters:
    data (list or np.array): The dataset.

    Returns:
    None (Prints skewness and kurtosis and displays a histogram).
    """
    # Compute skewness and kurtosis
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True returns excess kurtosis (normal = 0)

    # Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Excess Kurtosis: {kurtosis:.4f}")  # If close to 0, it's a normal distribution

    # Plot histogram
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='blue')
    plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label="Mean")
    plt.title("Histogram of Dataset (Checking Normality)")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()

# Generate a normal distribution dataset
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Normal(mean=50, std=15, size=1000)

# Compute and display skewness & kurtosis
calculate_skewness_kurtosis(data)
```
